## Using Method 1

```python
import pywt
```

```python
def w2d(img, mode='haar', level=1):
    imArray = img
    #Converting to grayscale
    imArray = cv2.cvtColor(imArray, cv2.COLOR_RGB2GRAY)
    imArray = np.float32(imArray)
    imArray /= 255
    
    #Computing coefficients
    coeffs = pywt.wavedec2(imArray, mode, level=level)
    
    #Process coefficients
    coeffs_h = list(coeffs)
    coeffs_h[0] *= 0
    
    #Reconstruction
    imArray_h = pywt.waverec2(coeffs_h, mode)
    imArray_h *= 255
    imArray_h = np.uint8(imArray_h)
    
    return imArray_h
```

```python
test_image = cv2.imread('test_image01.jpeg')
im_har = w2d(test_image, 'db1', 5)
plt.imshow(im_har, cmap='gray')
```

```python
#Creating data
data = []
labels = []
for index, celebrity in enumerate(celebrities):
    image_files = os.listdir(f'Images/{celebrity}')
    for image_file in image_files:
        image_path = f'Images/{celebrity}/{image_file}'
        image = cv2.imread(image_path)
        scaled_raw_img = cv2.resize(image, (32, 32))
        img_har = w2d(image, 'db1', 5)
        scaled_img_har = cv2.resize(img_har, (32, 32))
        #Stacking them together
        combined_img = np.vstack((scaled_raw_img.reshape(32*32*3, 1), scaled_img_har.reshape(32*32, 1)))
        data.append(combined_img)
        labels.append(index)
```

```python
print(len(data), len(data[0]), len(labels))
```

```python
X = np.array(data).reshape(len(data), len(data[0])).astype(float)
X.shape
```

```python
y = np.array(labels)
y.shape
```

<b> Model construction and training </b>

```python
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
```

```python
X_train, X_new, y_train, y_new = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_cv, X_test, y_cv, y_test = train_test_split(X_new, y_new, test_size=0.33, random_state=42)
```

<b> Hyper Parameter Tuning </b>

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
```

```python
model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto',probability=True),
        'params' : {
            'svc__C': [0.1, 1, 10, 100, 1000],
            'svc__kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'randomforestclassifier__n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(solver='liblinear',multi_class='auto'),
        'params': {
            'logisticregression__C': [1,5,10]
        }
    }
}
```

```python
scores = []
best_estimators = {}
import pandas as pd
for algo, mp in model_params.items():
    pipe = make_pipeline(StandardScaler(), mp['model'])
    clf =  GridSearchCV(pipe, mp['params'], cv=5, return_train_score=False)
    clf.fit(X_train, y_train)
    scores.append({
        'model': algo,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
    best_estimators[algo] = clf.best_estimator_
    
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```

## Using Method 2

```python
data = []
labels = []
for index, celebrity in enumerate(celebrities):
    image_files = os.listdir(f'Images/{celebrity}')
    for image_file in image_files:
        image_path = f'Images/{celebrity}/{image_file}'
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        image = cv2.resize(image, (128, 128)) / 255 # Resize and normalize (values between 0 and 1)
        data.append(image)
        labels.append(index)
```

```python
#Clearing RAM Memory
#import gc
#gc.collect()
```

```python
X = np.array(data)
y = np.array(labels)
```

```python
count_classes = len(celebrities)
count_classes
```

```python
#Checking labels and images
for i in range(0,len(X),1000):
    img = X[i]
    print(y[i], ': ', celebrities[y[i]])
    plt.imshow(img)
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.regularizers import l2, l1
```

```python
# Avoid OOM errors by setting GPU Memory Consumption Growth
gpus = tf.config.experimental.list_physical_devices('GPU')
for gpu in gpus: 
    tf.config.experimental.set_memory_growth(gpu, True)
```

```python
tf.config.list_physical_devices('GPU')
```

```python
model = Sequential()
model.add(Conv2D(128, (3, 3), 1, activation='relu', input_shape=(64, 64, 3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3, 3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(Dense(256, activation='relu', kernel_regularizer=l2(10)))
model.add(Dense(128, activation='relu'))
model.add(Dense(count_classes, activation='softmax'))
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_testing, y_train, y_testing = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_cv, X_test, y_cv, y_test = train_test_split(X_testing, y_testing, test_size=0.33, random_state=42)
```

<b> Training the Model </b>

```python
log_dir = 'logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir=log_dir)
print(tf.version.VERSION)
```

```python
hist = model.fit(X_train, y_train, batch_size=32, epochs=20, validation_data=(X_cv, y_cv), callbacks=[tensorboard_callback])
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color='teal', label='loss')
plt.plot(hist.history['val_loss'], color='orange', label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
def predict(image_file):
    test_image = cv2.imread(image_file)
    
    plt.imshow(test_image)
    plt.show()
    
    test_image = cv2.resize(test_image, (64, 64)) / 255.0  # Resize and normalize

    # Reshape the image to match the input shape of the model
    test_image = np.reshape(test_image, (1, 64, 64, 3))
    predictions = model.predict(test_image)
    predicted_label = np.argmax(predictions)
    celebrity_name = celebrities[predicted_label]

    print(f"The predicted celebrity is: {celebrity_name}")

image_file_name = 'test_image01.jpeg'
predict(image_file_name)
```
