```python
# Assigning strings to seasons
season_names = {1:"spring", 2:"summer", 3:"fall", 4:"winter"}
for key in season_names:
    df.loc[(df['season'] == key), 'season'] = season_names[key]
df
```

```python
df['season'].astype('category').value_counts()
```

```python
# Assigning strings to months
def map_months(x):
    return x.map({1:"Jan", 2:"Feb", 3:"Mar", 4:"Apr", 
                   5:"May", 6:"Jun", 7:"Jul", 8:"Aug", 
                   9:"Sept", 10:"Oct", 11:"Nov", 12:"Dec"})

df[['mnth']] = df[['mnth']].apply(map_months)    
```

```python
df
```

```python
# Assigning strings to weekdays

def map_weekday(x):
    return x.map({1:'Mon', 2:'Tues', 3:'Wed', 4:'Thurs', 5:'Fri', 6:'Sat', 0:'Sun'})
df[['weekday']] = df[['weekday']].apply(map_weekday)
```

```python
df
```

```python
# Assigning string to weather

def map_weather(x):
    return x.map({1:"A", 2:"B", 3:"C"})

df[['weathersit']] = df[['weathersit']].apply(map_weather) 
```

## Data prep

```python
df_categorical = df.select_dtypes(include=['object'])
df_categorical
```

```python
df_dummies = pd.get_dummies(df_categorical, drop_first=True).astype(int)
df_dummies
```

```python
# Drop categorical columns
df = df.drop(list(df_categorical.columns), axis = 1)
df
```

```python
# Concatenate dummy variables with dataset
df = pd.concat([df, df_dummies], axis = 1)
df
```

```python
df = df.drop(['instant', 'dteday'], axis=1)
df
```

## Model building

```python
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(0)
df_train, df_test = train_test_split(df, test_size=0.3, random_state=42)
```

```python
df_train
```

```python
# Feature Scaling
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
# Apply scalar function to non categorical columns
var = ['temp', 'hum', 'windspeed', 'casual', 'registered', 'cnt']
df_train[var] = scaler.fit_transform(df_train[var])
```

```python
df_train
```

```python
df_train.describe()
```

```python
# Checking correlation
plt.figure(figsize=(30,30))
sns.heatmap(df_train.corr(), annot = True, cmap = "YlGnBu")
plt.show()
```

```python
# Dividing into X and y
X_train = df_train.drop(['casual', 'registered'], axis=1)
y_train = df_train.pop('cnt')
```

```python
X_train
```

```python
y_train
```

```python
lm = LinearRegression()
lm.fit(X_train, y_train)
```

```python
from sklearn.feature_selection import RFE
```

```python
lm_new = LinearRegression()
rfe1 = RFE(estimator=lm_new, n_features_to_select=15)

rfe1.fit(X_train, y_train)
print(rfe1.support_)
print(rfe1.ranking_)
```

```python
col1 = X_train.columns[rfe1.support_]
col1
```

```python
rfe1.score(X_train, y_train)
```

```python
df_test[var] = scaler.fit_transform(df_test[var])
# Dividing into X and y
X_test = df_test.drop(['casual', 'registered'], axis=1)
y_test = df_test.pop('cnt')
```

```python
rfe1.score(X_test, y_test)
```
