Now we can see that features like Glucose, BoodPressure, Insulin, BMI, SkinThickness, Age might have 0 values. This behavior is obviously absurd and hence needs to be changed. 

One way to tackle this is to fill these fields with either mean or median values.

```python
# Replacing zero values with mean/median
cols_for_median = ['Glucose', 'BloodPressure', 'Age']
cols_for_mean = ['BMI', 'SkinThickness', 'Insulin']
for col in cols_for_median:
    df[col] = df[col].replace(0, df[col].median())
for col in cols_for_mean:
    df[col] = df[col].replace(0, df[col].mean())
df
```

```python
# We may drop some outliers from the data for better results

df = df[df['BMI']<55]
df = df[df['Insulin']<400]
df.describe()
```

```python
X = df.drop(['Outcome'], axis='columns')
X
```

```python
y = df['Outcome']
y
```

<b> Now we can test several models and choose the best one for our usage </b>

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
```

```python
#JSON for all models

model_params = {
    'svm': {
        'model': svm.SVC(gamma='auto'),
        'params' : {
            'C': [1,10,20],
            'kernel': ['rbf','linear']
        }  
    },
    'random_forest': {
        'model': RandomForestClassifier(),
        'params' : {
            'n_estimators': [1,5,10]
        }
    },
    'logistic_regression' : {
        'model': LogisticRegression(multi_class='auto'),
        'params': {
            'C': [1,5,10],
            'solver': ['lbfgs', 'liblinear']
        }
    },
    'mlp_classifier': {
        'model' : MLPClassifier(random_state=42),
        'params' : {
            'activation': ['logistic', 'relu'],
            'max_iter': [4500,8000]
        }
    }
}
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
```

```python
scaler = StandardScaler()
scores = []
for model_name, mp in model_params.items():
    # Create a pipeline with the scaler and the model
    pipeline = Pipeline([
        ('scaler', scaler),
        ('model', mp['model'])
    ])

    # Define the parameter grid including both scaler and model parameters
    param_grid = {
        'scaler__with_mean': [True, False]
    }
    param_grid.update({f'model__{key}': value for key, value in mp['params'].items()})  
    # Add model-specific parameters

    # Create the GridSearchCV object with the pipeline
    clf = GridSearchCV(pipeline, param_grid, cv=5, return_train_score=False)
    clf.fit(X, y)
    
    scores.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })

df_scores = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df_scores
```

```python
max_score_index = df_scores['best_score'].idxmax()
```

```python
for i in range(len(model_params)):
    if i==max_score_index:
        print("The best results are obtained by: ")
    print("Model Name:",df_scores.at[i,'model'])
    print("Model Score:",df_scores.at[i,'best_score'])
    print("Parameters:",df_scores.at[i,'best_params'])
    print("\n")
```

```python
#Hence we shall use Neural Networks as our model
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline

# We now define our final model
final_model = make_pipeline(StandardScaler(), MLPClassifier(activation='logistic', max_iter=4500))

scores = cross_val_score(final_model, X, y, cv=5)
print("Average Score of our model =", round(np.average(scores), 4)*100, "%")
```
