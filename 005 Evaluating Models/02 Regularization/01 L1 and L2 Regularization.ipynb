```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
dataset = pd.read_csv('melbourne.csv')
dataset.head(15)
```

```python
#Checking unique values in each field
dataset.nunique()
```

```python
dataset.shape
```

```python
#Dropping non relevant columns
cols_to_use = ['Suburb', 'Rooms', 'Type', 'Method', 'SellerG', 'Regionname', 'Propertycount', 
               'Distance', 'CouncilArea', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'Price']
dataset = dataset[cols_to_use]
dataset.head(15)
```

```python
#Checking for NA values
dataset.isna().sum()
```

```python
# Filling NA vals
cols_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)

# other continuous features can be imputed with mean for faster results since our focus is on Reducing overfitting
# using Lasso and Ridge Regression
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
#Drop NA values of Price, since it's our predictive variable and we can't impute it
dataset.dropna(inplace=True)
dataset.shape
```

<b> One hot encoding for categorical features </b>

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset.head()
```

```python
X = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=2)
```

<b> Let's train our Linear Regression Model on training dataset and check the accuracy on test set </b>

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X, train_y)
reg.score(test_X, test_y)
```

```python
reg.score(train_X, train_y)
```

We can see that the scores are very very low. Hence we need to change the model. 

Normal Regression is clearly overfitting the data, let's try other models

<b> Using Lasso (L1 Regularized) Regression Model (with modulus of paramters in regularisation term)</b>

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=1000, tol=0.1)
lasso_reg.fit(train_X, train_y)
```

```python
lasso_reg.score(test_X, test_y)
```

```python
lasso_reg.score(train_X, train_y)
```

We can see that the results have improved a bit. 

<b> Using Ridge (L2 Regularized) Regression Model (with squares of parameters in regularization term) </b>

```python
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(train_X, train_y)
```

```python
ridge_reg.score(test_X, test_y)
```

```python
ridge_reg.score(train_X, train_y)
```

<b> Linear Regression Model overfits. These results may not be that contrast but significant in most cases.Also that L1 & L2 Regularizations are used in Neural Networks too </b>
